```python
from ctgan import CTGANSynthesizer
from ctgan import load_demo
import pandas as pd
import warnings 

warnings.simplefilter("ignore")

data=pd.read_excel('../input/immunotherapy/Immunotherapy.xlsx')
data
```

We can decide its epoch and fit it accordingly. 

Where we need to pay attention is the similarity of the data so we constantly checking its values with .nunique() and unique() after training...

If we have discrete columns we need to fit it accordingly. 

**ctgan.fit(data, discrete_columns)**

```python
ctgan = CTGANSynthesizer(epochs=250)
ctgan.fit(data)

# Synthetic copy
samples = ctgan.sample(300)# How many rows we want
```
